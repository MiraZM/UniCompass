```python
## هسا الداتا الكبيرة اصعب لل ماتركس فاكتوريزيشن ف نقطة قوة النا عالاغلب
# ممكن تضيع الداتا ف الناتج مو نفس الاصلي
import numpy as np
#Prepping our data
print(new_only_marks_df.shape) # (610,1000)

avg_ratings = new_only_marks_df.mean(axis=1)
print(avg_ratings)
```

```python
#Prepping our data  (نغس اللي عملناه قبل لحتى نعبي السبارز)
only_marks_df_centered = new_only_marks_df.sub(avg_ratings, axis=0)
marks_df.fillna(0, inplace=True)
print(marks_df)
```

```python
#Applying SVD
from scipy.sparse.linalg import svds
U, sigma, Vt = svds(only_marks_df_centered,k=3) # بقدر اختار عدد ال ليتنت فيتشرز بس الديفولت 6
print(U.shape) #  (610,6)

print(Vt.shape) #(6,1000)

print(sigma) #بحس افضل نعرف كيف بتطلع سيجما

sigma = np.diag(sigma) # sigma has to be a diagonal matrix
print(sigma)
```

```python
#Calculating the product in Python
recalculated_ratings = np.dot(np.dot(U, sigma), Vt)
print(recalculated_ratings)
```

```python
#Add averages back #عشان تبطل سينترد
recalculated_ratings = recalculated_ratings + avg_ratings.values.reshape(-1, 1)

##compare original and new##

print(recalculated_ratings) # after doing matrix factorization on the original df

pd.DataFrame(marks_df) # original df
```
